```python
import pandas as pd
from tqdm import tqdm
import time
```

```python
data = pd.read_csv('series_dataset.csv')
```

```python
data
```

```python
# Fetch genres of all movies
genres_all_movies = [data.loc[i]['genres'].split(',') for i in tqdm(data.index)]
# Find the list of genres of all movies in alphabetical order
genres = sorted(list(set([item for sublist in tqdm(genres_all_movies) for item in sublist])))
genres
```

```python
movie_titles=[]
full_data=[]
```

```python
for i in tqdm(data.index):
    # Append movie title and the index of the movie
    movie_titles.append((data.loc[i]['originalTitle'].strip(), i,data.loc[i]['tconst'].strip()))
    # Add list of genres of the movies (1/0) to movie data
    movie_data = [1 if genre in data.loc[i]['genres'].split(',') else 0 for genre in genres]
    # Add IMDb score of the movie to the movie data
    movie_data.append(data.loc[i]['averageRating'])
    # Add record of movie to main data list
    full_data.append(movie_data)
```
